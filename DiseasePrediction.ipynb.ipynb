```python
# seperating the independent and the dependent variable

x = df_cleaned.drop("target", axis=1)
x.head()
```

```python
y = df_cleaned["target"]
y.head()
```

Tree-based models such as Decision Trees, Random Forests, and XGBoost are not sensitive to feature scaling

```python
# training my XGBoost model
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python

# Split data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
x_train.head(), x_test.head(), y_train.head(), y_test.head()
```

```python
# Initialize XGBoost model for classification (using the objective 'binary:logistic' for binary classification)
model = xgb.XGBClassifier(objective='binary:logistic', eval_metric='logloss')
```

```python
# Train the model
model.fit(x_train, y_train)
```

```python
from sklearn.model_selection import RandomizedSearchCV

# Setup the RandomizedSearchCV
param_dist = {
    'learning_rate': [0.01, 0.05, 0.1, 0.2, 0.3],  # Different learning rates
    'n_estimators': [50, 100, 200, 300],  # Number of trees (estimators)
    'max_depth': [3, 4, 5, 6, 7],  # Depth of each tree
    'min_child_weight': [1, 2, 3, 4],  # Minimum weight of children
    'subsample': [0.6, 0.7, 0.8, 0.9],  # Subsample ratio for each boosting round
    'colsample_bytree': [0.6, 0.7, 0.8, 1],  # Column sampling per tree
    'gamma': [0, 0.1, 0.2, 0.3, 0.4],  # Regularization parameter
}
random_search = RandomizedSearchCV(estimator=model, param_distributions=param_dist, n_iter=100, cv=3, verbose=2, random_state=42, n_jobs=-1)



# Fit the model
random_search.fit(x_train, y_train)

# Best parameters and best score
print("Best Parameters:", random_search.best_params_)
print("Best Score:", random_search.best_score_)
```

```python
# Best parameters from RandomizedSearchCV
best_params = random_search.best_params_
```

```python
# Initialize the XGBoost model with the best parameters
model = xgb.XGBClassifier(
    objective='binary:logistic',
    eval_metric='logloss',
    **best_params  # Unpacking the best parameters into the model
)
```

```python
# Now, fit the model with the training data
model.fit(x_train, y_train)
```

```python
# Predict on the test set
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, mean_squared_error
import numpy as np

# Make predictions on the test set
y_pred = model.predict(x_test)

# Get predicted probabilities for the positive class (1) for RMSE and AUC-ROC
y_prob = model.predict_proba(x_test)[:, 1]

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)

# Calculate F1-Score
f1 = f1_score(y_test, y_pred)

# Calculate precision
precision = precision_score(y_test, y_pred)

# Calculate recall
recall = recall_score(y_test, y_pred)

# Calculate RMSE (Root Mean Squared Error)
rmse = np.sqrt(mean_squared_error(y_test, y_prob))

# Print all the metrics
print(f"Accuracy: {accuracy:.4f}")
print(f"F1-Score: {f1:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"RMSE: {rmse:.4f}")
```

```python
# Evaluate the model's accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
from sklearn.metrics import confusion_matrix



# Generate confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Plot the confusion matrix using Seaborn
plt.figure(figsize=(6, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["No Heart Attack", "Heart Attack"], yticklabels=["No Heart Attack", "Heart Attack"])
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()
```

```python
# finding the most wrong prediction


# Create a DataFrame to compare true labels, predicted labels, and prediction probabilities
df_results = pd.DataFrame({
    'True Label': y_test,
    'Predicted Label': y_pred,
    'Prediction Probability': y_prob
})

# Find where the model made wrong predictions
df_results['Error'] = df_results['True Label'] != df_results['Predicted Label']

# Get the most wrong predictions (false positives and false negatives)
# Sort by the confidence (prediction probability) for the most "wrong" predictions
most_wrong_predictions = df_results[df_results['Error']].sort_values(by='Prediction Probability', ascending=False)

# Display the top 10 most wrong predictions
print(most_wrong_predictions.head(10))
```

```python
# model interpitation


import shap

# Initialize SHAP explainer for XGBoost
explainer = shap.Explainer(model)

# Get SHAP values for the test data
shap_values = explainer(x_test)

# Plot summary plot (Global explanation of feature importance)
shap.summary_plot(shap_values, x_test)

# Plot SHAP values for a single instance (Local explanation)
shap.initjs()
shap.force_plot(shap_values[0])
```

```python
%%writefile app.py
import pandas as pd
import streamlit as st
import xgboost as xgb
import numpy as np

# Load the trained model
model = xgb.XGBClassifier()
model.load_model("model.json")

st.title("XGBoost Model Deployment")
# Input features
st.sidebar.header('User Input Features')

def user_input_features():
    #model's features
    age = st.sidebar.number_input('age', value=0)
    sex = st.sidebar.number_input('sex', value=0)
    cp = st.sidebar.number_input('cp', value=0)
    trestbps = st.sidebar.number_input('trestbps', value=0)
    chol = st.sidebar.number_input('chol', value=0)
    fbs = st.sidebar.number_input('fbs', value=0)
    restecg = st.sidebar.number_input('restecg', value=0)
    thalach = st.sidebar.number_input('thalach', value=0)
    exang = st.sidebar.number_input('exang', value=0)
    oldpeak = st.sidebar.number_input('oldpeak', value=0)
    slope = st.sidebar.number_input('slope', value=0)
    ca = st.sidebar.number_input('ca', value=0)
    thal = st.sidebar.number_input('thal', value=0)

    return pd.DataFrame({'age': [age], 'sex': [sex], 'cp': [cp], 'trestbps': [trestbps],
                         'chol': [chol], 'fbs': [fbs], 'restecg': [restecg], 'thalach': [thalach],
                         'exang': [exang], 'oldpeak': [oldpeak], 'slope': [slope], 'ca': [ca],
                         'thal':[thal]})

# Getting user input
input_data = user_input_features()

# Predict using the model
prediction = model.predict(input_data)

# Display results
st.write('Prediction:', prediction)
```
